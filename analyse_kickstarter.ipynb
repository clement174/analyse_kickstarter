#### Les colonnes ont visiblement Ã©tÃ© dÃ©calÃ©es Ã  cause d'un character dans la catÃ©gorie name qui a Ã©tÃ© reconnu comme un sÃ©parateur et a donc dÃ©calÃ© le reste des elements.

#### 0.19% des lignes sont dÃ©calÃ©s d'un cran. On va essayer de les recaller. On va d'abord recomposer le nom complet du projet dans la colonne name afin de pouvoir supprimer l'info contenue dans les colonnes suivantes.  On va ensuite dÃ©caler les colonnes du nombre de cran nÃ©cessaire 

```python
# Recompose name of project
def get_full_name(mask_of_rows):
    rows_to_shift = ks_2016_df[mask_of_rows]
    new_name = rows_to_shift['name '] + rows_to_shift['category ']
    ks_2016_df.loc[mask_of_rows, ['name ']] = new_name
    return(ks_2016_df.loc[mask_of_rows, ['name ']])


# Move value column wise
def shift_columns(mask_row_to_shift, nb_col):
    for i_col in range(nb_col + 2, len(shift.columns)):
        # we move value of value_row to target_row
        value_col  = ks_2016_df[mask_row_to_shift][shift.columns[i_col]]
        
        # Replacing value in column (n) by value in column (n + nb_col)
        ks_2016_df.loc[mask_row_to_shift, [shift.columns[i_col - nb_col]]] = value_col
```

```python
mask_usd_null = ks_2016_df['usd pledged '].isnull()

country_dico = {
    "USD": {
        'country': 'US',
        'exchange_rate': 1
           },
    "GBP": {
        'country': 'GB',
        'exchange_rate': 0.770 
           },
    "CAD": {
        'country': 'CA',
        'exchange_rate': 1.379
           },
    "EUR": {
        'country': 'EU(non specified)',
        'exchange_rate': 0.940
           },
    "AUD": {
        'country': 'AU',
        'exchange_rate': 1.400
           },
    "SEK": {
        'country': 'SE',
        'exchange_rate': 8.910
           },
    "NZD": {
        'country': 'NZ',
        'exchange_rate': 1.494
           },
    "DKK": {
        'country': 'DK',
        'exchange_rate': 7.000
           },
    "NOK": {
        'country': 'NO',
        'exchange_rate': 8.745
           },
    "CHF": {
        'country': 'CH',
        'exchange_rate': 1.025
           }
}
```

###### Afin de remplir "usd pledged " on va rÃ©cuperer le montant dans pledged et le convertir

```python
ks_2016_df.loc[mask_usd_null, ['usd pledged ']] = usd_pledged_null['currency '].apply(lambda x: country_dico[x]['exchange_rate']) * usd_pledged_null['pledged '].astype(float)      
ks_2016_df.loc[mask_usd_null, ['usd pledged ']]
```
