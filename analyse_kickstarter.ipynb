#### Les rows qui ont des valeurs dans ces colonnes sont décalés. 
- Si une valeur dans unnamed: 13: décalage de 1 colonne
- Si une valeur dans unnamed: 14: décalage de 2 colonnes
- Si une valeur dans unnamed: 15: décalage de 3 colonnes
- Si une valeur dans unnamed: 16: décalage de 4 colonnes

```python
def get_percentage_shifted_rows(sub_df, nb_decimals):
    percentage_1 = round((len(sub_df) / len(ks_2016_df)) * 100, nb_decimals)
    return percentage_1
```

```python
print("{}% des lignes sont décalées d'un cran".format(get_percentage_shifted_rows(shift_1, 2)))
print("{}% des lignes sont décalées de deux crans".format(get_percentage_shifted_rows(shift_2, 4)))
print("{}% des lignes sont décalées de trois crans".format(get_percentage_shifted_rows(shift_3, 4)))
print("{}% des lignes sont décalées de quatres crans".format(get_percentage_shifted_rows(shift_4, 4)))
print("\n")
print("{}% des lignes sont décalées".format(get_percentage_shifted_rows(shift, 4)))
```

#### 0.19% des lignes sont décalés d'un cran. On va essayer de les recaller. On va d'abord recomposer le nom complet du projet dans la colonne name afin de pouvoir supprimer l'info contenue dans les colonnes suivantes.  On va ensuite décaler les colonnes du nombre de cran nécessaire 

```python
# Recompose name of project
def get_full_name(mask_of_rows):
    rows_to_shift = ks_2016_df[mask_of_rows]
    new_name = rows_to_shift['name '] + rows_to_shift['category ']
    ks_2016_df.loc[mask_of_rows, ['name ']] = new_name
    return(ks_2016_df.loc[mask_of_rows, ['name ']])


# Move value column wise
def shift_columns(mask_row_to_shift, nb_col):
    for i_col in range(nb_col + 2, len(shift.columns)):
        # we move value of value_row to target_row
        value_col  = ks_2016_df[mask_row_to_shift][shift.columns[i_col]]
        
        # Replacing value in column (n) by value in column (n + nb_col)
        ks_2016_df.loc[mask_row_to_shift, [shift.columns[i_col - nb_col]]] = value_col
```

###### Bingo, on va donc pouvoir traiter les deux problème avec le même masque

```python
usd_pledged_null['currency '].value_counts()
```

```python
mask_usd_null = ks_2016_df['usd pledged '].isnull()

country_dico = {
    "USD": {
        'country': 'US',
        'exchange_rate': 1
           },
    "GBP": {
        'country': 'GB',
        'exchange_rate': 0.770 
           },
    "CAD": {
        'country': 'CA',
        'exchange_rate': 1.379
           },
    "EUR": {
        'country': 'EU(non specified)',
        'exchange_rate': 0.940
           },
    "AUD": {
        'country': 'AU',
        'exchange_rate': 1.400
           },
    "SEK": {
        'country': 'SE',
        'exchange_rate': 8.910
           },
    "NZD": {
        'country': 'NZ',
        'exchange_rate': 1.494
           },
    "DKK": {
        'country': 'DK',
        'exchange_rate': 7.000
           },
    "NOK": {
        'country': 'NO',
        'exchange_rate': 8.745
           },
    "CHF": {
        'country': 'CH',
        'exchange_rate': 1.025
           }
}
```

###### Afin de remplir "usd pledged " on va récuperer le montant dans pledged et le convertir

```python
ks_2016_df.loc[mask_usd_null, ['usd pledged ']] = usd_pledged_null['currency '].apply(lambda x: country_dico[x]['exchange_rate']) * usd_pledged_null['pledged '].astype(float)      
ks_2016_df.loc[mask_usd_null, ['usd pledged ']]
```

```python
ks_2016_df[mask_usd_null]
```

###### On va remplir la colonne country en fonction de la currency.

```python
#ks_2016_df.loc[[usd_pledged_null.index], ['country ']] = country_dico[usd_pledged_null['currency ']]['country']
ks_2016_df.loc[mask_usd_null, ['country ']] = usd_pledged_null['currency '].apply(lambda x: country_dico[x]['country'])
ks_2016_df.loc[mask_usd_null, ['country ']]
```
